```python
def distmatrix(N):
    import csv
    DIST=[]
    with open('reuc_100', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=' ') #, quotechar='|')
        for i, row in enumerate(reader):
            if i >= 2+N:
                DIST.append(list(map(float, row)))
        return (DIST)
    return(999999999999999999)
            
```

```python
def closest(D, n):
    closest = []
    for i in range(n):
        dlist = [[D[i][j] ,j,i] for j in range(n) if j != i]#generates a list of neighbours sorted 
        dlist.sort()
        closest.append(dlist)
    return closest
```
